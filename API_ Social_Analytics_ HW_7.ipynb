```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(cons_key, cons_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
#Target media to mine 
target_user = ["@CBS", "@FoxNews", "@nytimes", "@CNN", "@BBC"]
# Dataset to store the results
Media_twitter = pd.DataFrame(columns=["Media", "Compound", "Positive", "Neutral", "Negative"], dtype= float)
Media_twitter
```

```python
Counter = 0 
# Loop through target users
for target in target_user:
    
    public_tweets = api.user_timeline(target, count=100)
    
    print("\n" + target + "\n")
       
        #Loops through tweets
    for tweet in public_tweets:
            # Run Vader Analysis on each tweet
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        print(tweet["text"])    
        # Amend the dataframe 
        Media_twitter.at[Counter, "Media"] = target
        Media_twitter.at[Counter, "Compound"] = compound
        Media_twitter.at[Counter, "Positive"] = pos
        Media_twitter.at[Counter, "Neutral"] = neu
        Media_twitter.at[Counter, "Negative"] = neg
        Counter+=1
```

```python
Media_twitter.head()
```

```python
plt.clf()
labels = ["CBS", "FoxNews", "nytimes", "CNN", "BBC"]
fig = plt.figure()
ax = plt.subplot()
for media, target in enumerate(target_user): 
    Media_twitter_df= Media_twitter[Media_twitter["Media"] ==  target]
    plt.scatter(np.arange(100,0,-1), Media_twitter_df["Compound"], marker = "o", label = labels[media])

ax.legend (loc = 'right', bbox_to_anchor=(1.28,.80), title ="Media Sources")
plt.title("Sentiment Analysis of Media Tweets")
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity") 
plt.style.use('ggplot')
plt.savefig("Sentiment Analysis Media Tweets.png")
plt.show()
```
